```python
import cv2
import matplotlib.pyplot as plt
```

## 1. Importing of the Image

```python
src = cv2.imread('/content/VIER PFOTEN_2015-04-27_010-1927x1333-1920x1328.jpg')
plt.imshow(src)
```

## 2. Edge-preserving Smoothing

We apply the Edge-preserving filter which is an image processing technique that smooths away noise or textures while retaining sharp edges

**sigma_s** controls how much the image is smoothed - the larger its value, the more smoothed the image gets, but it's also slower to compute.

**sigma_r** is important if you want to preserve edges while smoothing the image. Small sigma_r results in only very similar colors to be averaged (i.e. smoothed), while colors that differ much will stay intact.

```python
dst = cv2.edgePreservingFilter(src, flags=1, sigma_s=60, sigma_r=0.4)
plt.imshow(dst)
```

## 3. Gaussian Filter

```python
import numpy as np
kernel2 = np.ones((5, 5), np.float32) / 25
img = cv2.filter2D(src=src, ddepth=-1, kernel=kernel2)
plt.imshow(img)
```

## 4. Detail Enhancer Filter

This makes the image look Sharp

```python
dst = cv2.detailEnhance(src, sigma_s=10, sigma_r=0.15)
plt.imshow(dst)
```

## 5. Pencil Sketch Filter

This filter produces an output that looks like a pencil sketch. Image 1 shows the output when the input image is colored. The 2nd Image shows the output when the input image is grayscale

```python
dst_gray, dst_color = cv2.pencilSketch(src, sigma_s=60, sigma_r=0.07, shade_factor=0.05)
plt.imshow(dst_gray)
```

```python
plt.imshow(dst_color)
```

## 6. Stylization

```python
dst = cv2.stylization(src, sigma_s=60, sigma_r=0.07)
plt.imshow(dst)
```
